```python
import pandas as pd
import numpy as np
import re
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from statsmodels.stats.multicomp import pairwise_tukeyhsd


# Clean column names: replace spaces and special characters with underscores
df.columns = [re.sub(r'[^A-Za-z0-9]+', '_', col) for col in data.columns]

# Fill missing values in categorical columns with the mode
categorical_columns = df.select_dtypes(include=['object']).columns

for column in categorical_columns:
    df[column].fillna(df[column].mode()[0], inplace=True)

# Fill missing values in numerical columns with the mean
numerical_columns = df.select_dtypes(include=[np.number]).columns

for column in numerical_columns:
    df[column].fillna(df[column].mean(), inplace=True)

# Convert 'Last_Sem_CPI' to float
df['Last_Sem_CPI'] = df['Last_Sem_CPI'].astype(float)

# Build the formula for ANOVA with all categorical variables
formula = 'Last_Sem_CPI ~ ' + ' + '.join(categorical_columns)

# Perform ANOVA for all categorical variables
model_all = ols(formula, data=df).fit()
anova_results_all = anova_lm(model_all)

# Extract F-statistic and p-value
anova_results_df = anova_results_all[['F', 'PR(>F)']].reset_index()
anova_results_df.columns = ['Variable', 'F-statistic', 'p-value']

# Display the ANOVA results in a readable format
anova_results_df
```

```python
import pandas as pd
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Function to calculate Eta Squared for each variable from the ANOVA table
def eta_squared(aov_table):
    return aov_table['sum_sq'] / aov_table['sum_sq'].sum()

# Assuming anova_results_all is the DataFrame containing your ANOVA results
# and df is the DataFrame containing your data
eta_squared_results = eta_squared(anova_results_all)
eta_squared_df = pd.DataFrame(eta_squared_results, columns=['Eta Squared']).reset_index()
eta_squared_df.columns = ['Variable', 'Eta Squared']

# Function to perform post hoc tests for each categorical variable with significant results
def post_hoc_tests(df, categorical_column, response_variable):
    post_hoc_results = []
    if df[categorical_column].nunique() > 1:  # Ensure there is more than one group
        post_hoc = pairwise_tukeyhsd(endog=df[response_variable],
                                     groups=df[categorical_column],
                                     alpha=0.05)
        post_hoc_df = pd.DataFrame(post_hoc.summary().data[1:], columns=post_hoc.summary().data[0])
        post_hoc_results.append((categorical_column, post_hoc_df))
    return post_hoc_results

# Loop through categorical columns and apply post hoc tests if ANOVA p-value < 0.05
categorical_columns = eta_squared_df['Variable'].tolist()
post_hoc_dfs = []

for column in categorical_columns:
    if anova_results_all.loc[anova_results_all['Variable'] == column, 'PR(>F)'].values[0] < 0.05:
        post_hoc_results = post_hoc_tests(df, column, 'Last_Sem_CPI')
        for _, post_hoc_df in post_hoc_results:
            post_hoc_df['Variable'] = column  # Add the variable name to the results
            post_hoc_dfs.append(post_hoc_df)

# Combine all post hoc DataFrames into one
if post_hoc_dfs:
    all_post_hoc_df = pd.concat(post_hoc_dfs, ignore_index=True)
    # Display the combined DataFrame
    print(all_post_hoc_df)
else:
    print("No significant variables found for post hoc analysis.")

# Optional: Save the results to a CSV file
# all_post_hoc_df.to_csv('post_hoc_results.csv', index=False)
```

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score

def evaluate_models_with_preprocessing(dataset, target_column, num_columns):
    # Separate target variable and features
    X = dataset.drop(target_column, axis=1)
    y = dataset[target_column]

    # Ensure the number of columns selected is not greater than available columns
    available_columns = X.columns
    if num_columns > len(available_columns):
        raise ValueError(f"Number of columns requested ({num_columns}) exceeds the available columns ({len(available_columns)})")

    # Randomly select specified number of columns
    selected_columns = np.random.choice(available_columns, num_columns, replace=False)
    X = X[selected_columns]

    # Separate numerical and categorical columns
    numerical_cols = X.select_dtypes(include=['float', 'int']).columns
    categorical_cols = X.select_dtypes(include=['object']).columns

    # Handle missing values and one-hot encoding ONLY if there are columns of that type
    if len(numerical_cols) > 0:
        # Handle missing values for numerical features
        imputer_num = SimpleImputer(strategy='mean')
        X_num = imputer_num.fit_transform(X[numerical_cols])
        # Reshape y to a 2D array for compatibility with SimpleImputer
        y_imputed = imputer_num.fit_transform(y)
        y_imputed = y_imputed.reshape(-1, 1)
        # Flatten y back to 1D after imputation
        y = y_imputed
    else:
        X_num = np.array([]).reshape(len(X), 0)  # Create an empty array if no numerical columns

    if len(categorical_cols) > 0:
        # Handle missing values for categorical features
        imputer_cat = SimpleImputer(strategy='most_frequent')
        X_cat = imputer_cat.fit_transform(X[categorical_cols])

        # Convert all values in categorical columns to strings
        X_cat = X_cat.astype(str)

        # One-hot encode categorical features
        encoder = OneHotEncoder(sparse=False, drop='first', handle_unknown='ignore')
        X_cat_encoded = encoder.fit_transform(X_cat)
    else:
        X_cat_encoded = np.array([]).reshape(len(X), 0)  # Create an empty array if no categorical columns

    # Combine numerical and encoded categorical features
    X_processed = np.hstack((X_num, X_cat_encoded))

    # Standardize data
    scaler = StandardScaler()
    X_processed = scaler.fit_transform(X_processed)

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.3, random_state=42)

    # Initialize models
    models = {
        'Linear Regression': LinearRegression(),
        'Random Forest': RandomForestRegressor(),
        'Support Vector Regressor': SVR(),
        'Decision Tree Regressor': DecisionTreeRegressor(),
        'K-Neighbors Regressor': KNeighborsRegressor()
    }

    # Train and evaluate each model
    r2_scores = {}
    for model_name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        r2_scores[model_name] = r2_score(y_test, y_pred)

    return r2_scores
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_cpi_distribution(data, hue_column):
    """
    Plots histograms with KDE for the distribution of 'Last Sem CPI'
    differentiated by a given column.

    Parameters:
    data (DataFrame): The dataset to plot.
    hue_column (str): The column name to differentiate by (e.g., 'Category' or 'Parental Education').
    """
    # Create subplots
    fig, axs = plt.subplots(1, 3, figsize=(25, 6))

    # Plot the first histogram with KDE, differentiated by hue_column (Overall)
    plt.subplot(131)
    sns.histplot(data=data, x='Last Sem CPI', kde=True, hue=hue_column)
    plt.title(f'Distribution of Last Sem CPI by {hue_column} (Overall)')
    plt.xlabel('Last Sem CPI')
    plt.ylabel('Frequency')

    # Plot the second histogram with KDE, differentiated by hue_column (Female)
    plt.subplot(132)
    female_data = data[data.Gender == 'Female']
    sns.histplot(data=female_data, x='Last Sem CPI', kde=True, hue=hue_column)
    plt.title(f'Distribution of Last Sem CPI by {hue_column} (Female)')
    plt.xlabel('Last Sem CPI')
    plt.ylabel('Frequency')

    # Plot the third histogram with KDE, differentiated by hue_column (Male)
    plt.subplot(133)
    male_data = data[data.Gender == 'Male']
    sns.histplot(data=male_data, x='Last Sem CPI', kde=True, hue=hue_column)
    plt.title(f'Distribution of Last Sem CPI by {hue_column} (Male)')
    plt.xlabel('Last Sem CPI')
    plt.ylabel('Frequency')

    # Display the subplots
    plt.tight_layout()
    plt.show()
    print(data[hue_column].value_counts())
```
